```python
from nilearn import datasets

# Download the Schaefer 2018 atlas with 100 parcels per hemisphere
schaefer_atlas = datasets.fetch_atlas_schaefer_2018(n_rois=100, resolution_mm=2)

# Print the path to the atlas files
print("Schaefer Atlas path:", schaefer_atlas['maps'])

# Download the Schaefer 2018 atlas with 100 parcels per hemisphere and 17 networks
schaefer_atlas = datasets.fetch_atlas_schaefer_2018(n_rois=100, yeo_networks=17, resolution_mm=2)

# Print the path to the atlas files
print("Schaefer 17-Network Atlas path:", schaefer_atlas['maps'])

# Download Yeo 2011 parcellation
yeo_atlas = datasets.fetch_atlas_yeo_2011(data_dir=None, url=None, resume=True, verbose=1)

# Path to the 7-network parcellation NIfTI file
yeo_7_nifti_path = yeo_atlas.thick_7

# Path to the 17-network parcellation NIfTI file
yeo_17_nifti_path = yeo_atlas.thick_17

print("7-Network Yeo parcellation NIfTI file path:", yeo_7_nifti_path)
print("17-Network Yeo parcellation NIfTI file path:", yeo_17_nifti_path)

# Write the volume of each network (in number of voxels)
import nibabel as nib
import numpy as np
    
# Load the 7-network parcellation
yeo_7_img = nib.load(yeo_7_nifti_path)
yeo_7_data = yeo_7_img.get_fdata()

# Load the 17-network parcellation
yeo_17_img = nib.load(yeo_17_nifti_path)
yeo_17_data = yeo_17_img.get_fdata()

# Get the unique values in the 7-network parcellation
unique_values_7 = np.unique(yeo_7_data)

# Get the unique values in the 17-network parcellation
unique_values_17 = np.unique(yeo_17_data)

# Count the number of voxels in each network for the 7-network parcellation
network_voxels_7 = {int(value): np.sum(yeo_7_data == value) for value in unique_values_7}
    
# Count the number of voxels in each network for the 17-network parcellation
network_voxels_17 = {int(value): np.sum(yeo_17_data == value) for value in unique_values_17}

# Print the number of voxels in each network for the 7-network parcellation
print("Number of voxels in each network for the 7-network parcellation:")
non_zero_networks_volume_list = []
for network, voxels in network_voxels_7.items():
    if network != 0:
        non_zero_networks_volume_list.append(voxels)
    # print(f"Network {network}: {voxels} voxels")
print(f"Average number of voxels in non-zero networks: {np.mean(non_zero_networks_volume_list)}")
# print the non_zero_networks_volume_list on one line
print(non_zero_networks_volume_list)
# save txt in the tests folder of this package (this notebook is in notebooks folder of bcblib, tests folder is in bcblib)
np.savetxt('../tests/yeo_7_networks_voxels.txt', non_zero_networks_volume_list, fmt='%d')
    
# Print the number of voxels in each network for the 17-network parcellation
print("Number of voxels in each network for the 17-network parcellation:")
non_zero_networks_volume_list = []
for network, voxels in network_voxels_17.items():
    if network != 0:
        non_zero_networks_volume_list.append(voxels)
    # print(f"Network {network}: {voxels} voxels")
print(f"Average number of voxels in non-zero networks: {np.mean(non_zero_networks_volume_list)}")
# print the non_zero_networks_volume_list on one line
print(non_zero_networks_volume_list)
# save txt in the tests folder of this package (this notebook is in notebooks folder of bcblib, tests folder is in bcblib)
np.savetxt('../tests/yeo_17_networks_voxels.txt', non_zero_networks_volume_list, fmt='%d')
```

```python
custom1 = "/gamma/Dropbox (GIN)/parcitron_tests/output_directory/Compactor_custom_sizes_noncontig_17/custom_sizes.nii.gz"

from nilearn.image import load_img
from nilearn.regions import connected_regions
import numpy as np

# Load the NIfTI image
nii_img = load_img(custom1)

# Get the data from the image as a numpy array
data = nii_img.get_fdata()

# Get the unique cluster labels (excluding 0 if that's background)
cluster_labels = np.unique(data)
cluster_labels = cluster_labels[cluster_labels != 0]

# Check contiguity for each cluster
contiguity_results = {}
for cluster in cluster_labels:
    # Create a binary mask for the current cluster
    cluster_mask = (data == cluster)
    cluster_nii_img = nib.Nifti1Image(cluster_mask.astype(float), affine=nii_img.affine)
    
    # Use connected_regions to find connected components
    connected_img, index = connected_regions(cluster_nii_img, min_region_size=1, extract_type='connected_components', smoothing_fwhm=0)
    
    # Check if there's more than one connected component
    is_contiguous = len(index) == 1
    contiguity_results[cluster] = is_contiguous

contiguity_results


```
